## JSON ETL spec based on orientdb-etl

```python
example_etl = {
    "extractor": {
        "networkx": {
            "class": "subgraph",
            "node_type_attr": "type",
            "edge_type_attr": "type",
            "traversal": [
                {"node": {"type": "Person", "alias": "p1"}},
                {"edge": {}},
                {"node": {"alias": "wild"}},
                {"edge": {}},
                {"node": {"type": "Person", "alias": "p2"}}
            ]
        }
    },
    "transformers": [
        {
            "project": {
                "method": {"jaccard": {"over": ["Institution", "City"]}},
                "pattern": [
                    {"node": {"alias": "p1"}},
                    {"edge": {}},
                    {"node": {"alias": "p2"}}
                ],
                "set": [
                    {
                        "alias": "NEW",
                        "key": "name",
                        "value":"",
                        "value_lookup": "wild.name"
                    }
                ],
                "delete": {"alias": ["wild"]}
            }
        }
    ],
    "loader": {
        "networkx": {}
    }
}
```

```python
subgraph = px.execute_etl(g, example_etl)
px.draw_simple_graph(subgraph)
```
